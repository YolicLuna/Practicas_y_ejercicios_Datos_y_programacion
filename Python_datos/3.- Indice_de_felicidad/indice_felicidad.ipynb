```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargamos los datos de cada año.

df_2015 = pd.read_csv('Data_bases/2015.csv')
df_2016 = pd.read_csv('Data_bases/2016.csv')
df_2017 = pd.read_csv('Data_bases/2017.csv')
df_2018 = pd.read_csv('Data_bases/2018.csv')
df_2019 = pd.read_csv('Data_bases/2019.csv')

# Mostramos las primeras 2 filas de cada DataFrame para verificar la carga de datos.

display(df_2015.head(2), df_2016.head(2), df_2017.head(2), df_2018.head(2), df_2019.head(2))
```

```python
# Verificamos las columnas de cada DataFrame para identificar posibles diferencias.

print('Columnas de 2015:', df_2015.columns)
print('Columnas de 2016:', df_2016.columns)
print('Columnas de 2017:', df_2017.columns)
print('Columnas de 2018:', df_2018.columns)
print('Columnas de 2019:', df_2019.columns)
```

```python
# Renombramos columnas para unificar los nombres y facilitar el análisis.

df_2015.rename(columns={
    'Economy (GDP per Capita)': 'GDP per Capita',
    'Health (Life Expectancy)': 'Healthy Life Expectancy',
    'Trust (Government Corruption)': 'Perceptions of Corruption',
    'Family': 'Social Support',
    'Freedom': 'Freedom to Make Life Choices'
}, inplace=True)

df_2016.rename(columns={
    'Economy (GDP per Capita)': 'GDP per Capita',
    'Health (Life Expectancy)': 'Healthy Life Expectancy',
    'Trust (Government Corruption)': 'Perceptions of Corruption',
    'Family': 'Social Support',
    'Freedom': 'Freedom to Make Life Choices'
}, inplace=True)

df_2017.rename(columns={
    'Happiness.Score': 'Happiness Score',
    'Economy..GDP.per.Capita.': 'GDP per Capita',
    'Health..Life.Expectancy.': 'Healthy Life Expectancy',
    'Trust..Government.Corruption.':'Perceptions of Corruption',
    'Family':'Social Support',
    'Freedom':'Freedom to Make Life Choices'
}, inplace=True)

df_2018.rename(columns={
    'Country or region': 'Country',
    'Score': 'Happiness Score',
    'GDP per capita': 'GDP per Capita',
    'Social support': 'Social Support',
    'Healthy life expectancy': 'Healthy Life Expectancy',
    'Freedom to make life choices': 'Freedom to Make Life Choices',
    'Perceptions of corruption': 'Perceptions of Corruption',
    'Overall rank': 'Happiness Rank'
}, inplace=True)

df_2019.rename(columns={
    'Country or region': 'Country',
    'Score': 'Happiness Score',
    'GDP per capita': 'GDP per Capita',
    'Social support': 'Social Support',
    'Healthy life expectancy':'Healthy Life Expectancy',
    'Freedom to make life choices':'Freedom to Make Life Choices',
    'Perceptions of corruption':'Perceptions of Corruption',
    'Overall rank':'Happiness Rank'
}, inplace=True)
```

```python
# Agregamos una columna de año para cada DataFrame para facilitar el análisis temporal.

df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019
```

```python
# Concatenamos todos los DataFrames en uno solo para facilitar el análisis conjunto.

df_all_years = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)
display(df_all_years.head())
```

```python
# Verificamos la estructura del DataFrame combinado para asegurarnos de que todo esté correcto.

display(df_all_years.info())
```

```python
# Mostramos estadísticas descriptivas para entender mejor la distribución de los datos.

display(df_all_years.describe())
```

```python
# Verificamos si hay valores nulos en el DataFrame combinado.

display(df_all_years.isnull().sum())
```

```python
# Visualizamos los valores faltantes con un mapa de calor para identificar patrones de datos faltantes.

plt.figure(figsize=(10, 8))
sns.heatmap(df_all_years.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de calor de valores faltantes')
plt.show()
```

# Mapa de calor de valores faltantes

Esta gráfica muestra qué datos están completos y cuáles faltan en el dataset.
- Los espacios en color oscuro representan valores presentes.
- Los espacios en color amarillo indican datos faltantes.

### Interpretación:
Se observa que la mayoría de las variables están completas, pero algunas columnas tienen varios valores faltantes, lo cual indica que fue necesario limpiar o tratar esos datos antes del análisis.

```python
# Eliminamos columnas que no aportan información relevante para el análisis o que tienen muchos valores faltantes.

df_cleaned = df_all_years.drop(columns=[
    'Region', 'Standard Error', 'Lower Confidence Interval', 'Upper Confidence Interval',
    'Happiness Rank', 'Happiness.Rank', 'Whisker.high', 'Whisker.low','Dystopia Residual', 'Dystopia.Residual'
])
```

```python
# Eliminamos filas con valores nulos en las columnas clave para el análisis.

df_cleaned = df_cleaned.dropna(subset=[
    'Happiness Score', 'GDP per Capita', 'Social Support', 'Healthy Life Expectancy',
    'Freedom to Make Life Choices', 'Generosity', 'Perceptions of Corruption'
])
```

```python
# Verificamos nuevamente los valores nulos después de la limpieza.

print('Valores nulos restantes por columna:', df_cleaned.isnull().sum())
```

```python
# Mostramos el DataFrame limpio para verificar los cambios realizados.

display(df_cleaned)
```

```python
# Verificamos los años presentes en el DataFrame final para asegurarnos de que todos los años estén representados.

print('Años presentes en el DataFrame final:', df_cleaned['Year'].unique())
```

```python
# Analizamos la correlación entre las variables clave utilizando un mapa de calor.

variables = ['Happiness Score', 'GDP per Capita', 'Social Support', 'Healthy Life Expectancy',
             'Freedom to Make Life Choices', 'Generosity', 'Perceptions of Corruption']
correlation_matrix = df_cleaned[variables].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlacion entre Variables Clave')
plt.show()
```

# Matriz de correlación entre variables clave
Esta gráfica muestra qué tan relacionadas están las variables entre sí, con valores entre -1 y 1.

### Interpretación:
El Happiness Score tiene alta correlación con GDP per Capita (0.79) y Healthy Life Expectancy (0.74).
También tiene relación moderada con Social Support.
Generosity tiene poca o casi nula correlación con la felicidad.
En general, factores económicos y de salud influyen más en la felicidad que otros factores.

```python
# Analizamos la evolución del puntaje de felicidad promedio a lo largo de los años.

happiness_by_year = df_cleaned.groupby('Year')['Happiness Score'].mean()
display(happiness_by_year)
```

```python
# Visualizamos la evolución del Happiness Score promedio por año utilizando un gráfico de barras.

plt.figure(figsize=(10, 8))
happiness_by_year.plot(kind='bar', color='skyblue')
plt.title('Evolucion del Happiness Score Promedio por Año')
plt.xlabel('Año')
plt.ylabel('Happiness Score Promedio')
plt.xticks(rotation=0)
plt.show()
```

# Evolución del Happiness Score promedio por año (barras)
Esta gráfica muestra el promedio de felicidad por año de 2015 a 2019 usando barras.
### Interpretación:
El nivel de felicidad se mantiene bastante estable a lo largo de los años, con pequeñas variaciones, sin cambios drásticos.

```python
# Visualizamos la evolución del Happiness Score promedio por año utilizando un gráfico de líneas.

plt.figure(figsize=(10, 8))
plt.plot(happiness_by_year, marker='o', linestyle='-',  color='skyblue')
plt.title('Evolucion del Happiness Score Promedio por Año')
plt.xlabel('Año')
plt.ylabel('Happiness Score Promedio')
plt.grid(True)
plt.xticks(happiness_by_year.index)
plt.show()
```

# Evolución del Happiness Score promedio por año (línea)
Es la misma información que la anterior, pero representada como una línea para ver mejor la tendencia.
### Interpretación:
Se observa una ligera caída en 2017 y una recuperación en los años siguientes, alcanzando su punto más alto en 2019.

```python
# Identificamos los países más felices y menos felices en el año 2019.

df_2019_clean = df_cleaned[df_cleaned['Year'] == 2019]
top_5_happiest = df_2019_clean.nlargest(5, 'Happiness Score')[['Country', 'Happiness Score']]
bottom_5_least_happy = df_2019_clean.nsmallest(5, 'Happiness Score')[['Country', 'Happiness Score']]

print('Paises mas felices en 2019:')
display(top_5_happiest)

print('Paises menos felices en 2019:')
display(bottom_5_least_happy)
```

```python
# Visualizamos los países más felices en 2019 utilizando un gráfico de barras.

plt.figure(figsize=(10, 8))
sns.barplot(x='Happiness Score', y='Country', data=top_5_happiest, palette='Blues_d')
plt.title('Top 5 Paises Mas Felices en 2019')
plt.show()
```

# Top 5 países más felices en 2019
Esta gráfica muestra los 5 países con mayor nivel de felicidad en 2019.

### Interpretación:
Los países más felices son:
- Finlandia.
- Dinamarca.
- Noruega.
- Islandia.
- Países Bajos.

Todos tienen niveles de felicidad muy altos y similares, lo que sugiere condiciones de vida comparables entre ellos.

```python
# Visualizamos los países menos felices en 2019 utilizando un gráfico de barras.

plt.figure(figsize=(10, 8))
sns.barplot(x='Happiness Score', y='Country', data=bottom_5_least_happy, palette='Reds_d')
plt.title('Top 5 Paises Menos Felices en 2019')
plt.show()
```

# Top 5 países menos felices en 2019

Esta gráfica muestra los 5 países con menor nivel de felicidad en 2019.

### Interpretación:
Los países menos felices son:
- Sudán del Sur
- República Centroafricana
- Afganistán
- Tanzania
- Ruanda

Todos presentan niveles de felicidad considerablemente bajos, lo que puede estar relacionado con factores como conflictos, pobreza o baja calidad de vida.

```python
# Analizamos la distribución del Happiness Score utilizando un diagrama de cajas para identificar posibles valores atípicos.

plt.figure(figsize=(10, 8))
sns.boxplot(y='Happiness Score', data=df_cleaned)
plt.title('Diagrama de Cajas para Happiness Score')
plt.ylabel('Happiness Score')
plt.show()
```

# Diagrama de cajas (Boxplot) del Happiness Score

Esta gráfica muestra la distribución del nivel de felicidad, incluyendo mediana, rango y posibles valores extremos.

### Interpretación:
La mayoría de los valores se concentran entre aproximadamente 4.5 y 6.5.
La mediana está alrededor de 5.3.
Existen valores más bajos y más altos, pero no son predominantes.

Esto indica que la mayoría de los países tienen niveles de felicidad moderados.

```python
# Analizamos la distribución del Happiness Score utilizando un histograma para visualizar la frecuencia de los puntajes.

plt.figure(figsize=(10, 8))
plt.hist(df_cleaned['Happiness Score'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribucion del Happiness Score (Histograma)')
plt.xlabel('Happiness Score')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

# Distribución del Happiness Score (Histograma)

Esta gráfica muestra cuántos países hay en cada rango de felicidad.

### Interpretación:
La mayor concentración de países está entre valores de 4 y 6.5.
Hay pocos países con niveles muy bajos o muy altos.

Esto confirma que la felicidad global se distribuye principalmente en niveles medios.

```python
# Analizamos la relación entre el GDP per Capita y el Happiness Score utilizando un gráfico de dispersión.

plt.figure(figsize=(10, 8))
sns.scatterplot(x='GDP per Capita', y='Happiness Score', data=df_cleaned)
plt.title('Relacion entre GDP per Capita y Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

# Relación entre GDP per Capita y Happiness Score

Esta gráfica muestra la relación entre el ingreso por persona y el nivel de felicidad.

### Interpretación:
Existe una relación positiva: a mayor GDP per capita, mayor felicidad.
Los puntos siguen una tendencia ascendente clara.

Esto sugiere que los países con mejores condiciones económicas tienden a tener mayor nivel de felicidad.

```python
# Calculamos la correlación entre el GDP per Capita y el Happiness Score para cuantificar la relación entre estas dos variables.

gdp_happiness_corr = df_cleaned['GDP per Capita'].corr(df_cleaned['Happiness Score'])
print(f'Correlacion entre GDP per Capita y Happiness Score: {gdp_happiness_corr:.2f}')
```

# Conclusión

- El análisis del índice de felicidad muestra que el bienestar de los países está fuertemente influenciado por factores económicos, sociales y de salud. Variables como el PIB per cápita, la esperanza de vida y el apoyo social presentan una relación significativa con el nivel de felicidad, lo que indica que mejores condiciones de vida contribuyen directamente al bienestar de la población.

- Se observa que la felicidad global se mantiene relativamente estable a lo largo del tiempo, con ligeras variaciones entre años. La mayoría de los países se concentran en niveles medios de felicidad, mientras que los extremos (países muy felices o muy poco felices) son menos frecuentes.

- También destaca una clara diferencia entre los países con mayores y menores niveles de felicidad, evidenciando desigualdades importantes a nivel mundial. Los países más felices tienden a compartir características como estabilidad económica, sistemas de salud sólidos y apoyo social, mientras que los menos felices suelen enfrentar mayores dificultades estructurales.

- En conjunto, este análisis confirma que la felicidad no depende de un solo factor, sino de la combinación de múltiples variables que reflejan la calidad de vida de una sociedad.
